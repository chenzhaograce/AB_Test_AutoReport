# Binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
#import binomial distribution
from scipy.stats import binom
#import normal distribution
from scipy.stats import norm
#import chi-square distribution
from scipy.stats import chi2
```

## The binomial distribution is used when:

- n independent outcomes
- Outcome is binary, often generically denoted as “success” or “failure”
- p(success) remains constant throughout the experiment

### Yes/No type outcomes
- Incredibly common


```python
#write a function to calculate the probability of getting 2 heads in 3 tosses of a coin
def prob_2_heads_in_3_tosses():
    #get the probability of getting 2 heads in 3 tosses of a coin
    prob_2_heads_in_3_tosses = binom.pmf(2,3,0.5).round(2)    #print the probability
    print("The probability of getting 2 heads in 3 tosses of a coin is", prob_2_heads_in_3_tosses)
    #plot the probability distribution
    x = np.arange(0,4)
    y = binom.pmf(x,3,0.5)
    plt.plot(x,y)
    plt.xlabel("Number of heads")
    plt.ylabel("Probability")
    plt.title("Probability distribution of getting 2 heads in 3 tosses of a coin")
    plt.show()
```

```python
#print the probability of getting 2 heads in 3 tosses of a coin
prob_2_heads_in_3_tosses()
```

```python
#Find the probability that exactly 3 of the 10 individuals make a purchase with a probability of success of 0.2
def prob_3_of_10_individuals_make_a_purchase():
    #get the probability that exactly 3 of the 10 individuals make a purchase with a probability of success of 0.2 to make a purchase
    prob_3_of_10_individuals_make_a_purchase = binom.pmf(3,10,0.2).round(2)    #print the probability
    print("The probability that exactly 3 of the 10 individuals make a purchase with a probability of success of 0.2 to make a purchase is", prob_3_of_10_individuals_make_a_purchase)
    #plot the probability distribution
    x = np.arange(0,11)
    y = binom.pmf(x,10,0.2)
    plt.plot(x,y)
    plt.xlabel("Number of individuals")
    plt.ylabel("Probability")
    plt.title("Probability distribution of exactly 3 of the 10 individuals make a purchase with a probability of success of 0.2")
    plt.show()
```

```python
#print the result
prob_3_of_10_individuals_make_a_purchase()
```
